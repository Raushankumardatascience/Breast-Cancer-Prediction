```python
#Here "M" stands for malignant, indicating cancerous tumors.
# while "B" stands for benign, indicating non-cancerous tumors.

"""
M (Malignant):

Represents cancerous tumors.
Indicates that the tumor has the potential to grow and spread to other parts of the body.
B (Benign):

Represents non-cancerous tumors.
Indicates that the tumor is not harmful and does not spread to other tissues.
"""
```

```python
#Converting M and B into Integer by LabelEncoding
#import LabelEncoding
from sklearn.preprocessing import LabelEncoder
encode = LabelEncoder()

df["diagnosis"] = encode.fit_transform(df["diagnosis"])# M is replaced  by 1 and B is replaced by 0.
```

```python
correlation_value = df[["diagnosis", "id"]].corr()
print(correlation_value)
 #Here id doesot distribute as such in the model
```

```python
#Here corr()value  of id and diagnosis value is low we can drop it
df = df.drop("id", axis = 1)
```

```python
#Extracrt X and Y
x = df.drop(columns = "diagnosis", axis = 1)
y = df["diagnosis"]

# import train_test_split for model_traning
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 42)

#Import LogisticRegression alogorithm
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()


#For hyperparameter tunnig
from sklearn.model_selection import GridSearchCV
parms_grid = {'C': [1,10,20,30,40],
        'penalty': ('l1','l2','elasticnet')}

#Fit the Model
grid = GridSearchCV(model, parms_grid, cv = 5, verbose = 2)
grid.fit(x_train, y_train)

model.fit(x_train, y_train)

#import metrics
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

x_test.shape, x_train.shape, y_test.shape, y_train.shape
```

```python
grid.best_estimator_
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
#Prediction after using Hyperparameters(best_estimator)

y_predict_hyp = grid.best_estimator_.predict(x_test)
```

```python
accuracy_score(y_test, y_predict_hyp)
```

```python
#prediction with normal parameters
y_predict = model.predict(x_test)
```

```python
accuracy_score(y_test, y_predict)
```

```python
print(classification_report(y_test, y_predict))
```

```python
print(confusion_matrix(y_test, y_predict))
```
